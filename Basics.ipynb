### Mutating & Non-Mutating Functions
By convention, functions followed by `!` alter their contents and functions lacking `!` do not.

```julia
C = [3, 5, 4]
```

```julia
sort(C) #sorts
```

```julia
C #C remains the same
```

```julia
D = [3,5,4]
```

```julia
sort!(D) #sorts and alters D in place
#= this works somewhat like
D = sort(D)
=#
```

```julia
D
```

To add/remove element from an array, mutating functions `push!()` and `pop!()` were used.

### Broadcasting
`function.(argument)` -- to call a function as broadcasting

```julia
f(A) #squares the matrix
```

```julia
broadcast(f, A) #squares individual elements
```

```julia
f.(A) #squares individiual elements
```

```julia
broadcast(x -> x + 2 * f(x) / x, A)
#= f(x) = x^2
2*f(x)/x = 2*x
x + 2 * f(x) = 3*x
=> this command multiplies each element of the matrix A by 3 
=#
```

```julia
A .+ 2 .* f.(A) ./A #works the same as last cell
```

#### GR Backend
`Plots.jl` allows you to seamlessly change backends(libraries that actually do the drawing)<br>
`GR` is the default backend.

```julia
gr()
```

```julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17]

plot(numpirates, globaltemperatures, label="line")  
scatter!(numpirates, globaltemperatures, label ="scatter") #it is used as a mutating function, as the plot created by plot is edited. if used without !, a new plot(scatter) will be created and first plot will be overridden.
```

```julia
scatter(numpirates, globaltemperatures, label = "scatter")
plot!(numpirates, globaltemperatures, label = "line") #plotting order is reversed.
```

```julia
#Add more things
xlabel!("Number of pirates")
ylabel!("Global Temperature")
title!("Pirate Population & Global Warming")
```

This data should have been read as:<br>
Number of pirates decreased with time(years) and global temperature rose.<br>
So with the current plot, it has to be read right to left.<br>

`This can be rectified by flipping x`

```julia
xflip!()
```

```julia
time_comparison = Dict()
a = rand(10^7)
```

```julia
pysum = pybuiltin("sum")
py_bench = @benchmark pysum(a)
```

```julia
minimum(py_bench.times)
```

```julia
time_comparison["python-builtin"] = minimum(py_bench.times)/1e6
```

```julia
# using Conda
# Conda.add("numpy")
```

```julia
@pyimport numpy as np
```

```julia
py_np_bench = @benchmark np.sum(a)
```

```julia
minimum(py_np_bench.times)
```

```julia
time_comparison["python-numpy"] = minimum(py_np_bench.times)/1e6
```

```julia
julia_bench = @benchmark sum(a)
```

```julia
time_comparison["julia-builtin"] = minimum(julia_bench.times)/1e6
```

```julia
x = []
y = []
for (k,v) in time_comparison
    push!(x ,k)
    push!(y, v)
end
println(x)
println(y)
```

```julia
# using Conda
# Conda.add("matplotlib")
```

```julia
@pyimport matplotlib.pyplot as plt
```

```julia
plt.bar(x, y)
plt.show()
```
